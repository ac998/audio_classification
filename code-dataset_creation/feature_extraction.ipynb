```python
import numpy as np
import librosa as lbr
import os
import matplotlib.pyplot as plt
```

```python
n_mfcc = 26
```

```python
# machine sound
path = "machine_data_final_750"
mach_files = os.listdir(path)
machine_features = np.zeros((len(mach_files), n_mfcc))

for i, fname in enumerate(mach_files):
    wave,sr = lbr.load(os.path.join(path, fname))
    mfcc = np.mean(lbr.feature.mfcc(y=wave, sr=sr, n_mfcc=n_mfcc), axis=1)
    machine_features[i, : ] = mfcc
    print(sr, i)
```

```python
print(machine_features.shape)
print(len(mach_files))
```

```python
# music sound
path = "music_data_final_750"
mus_files = os.listdir(path)
music_features = np.zeros((len(mus_files), n_mfcc))

for i, fname in enumerate(mus_files):
    wave,sr = lbr.load(os.path.join(path, fname))
    mfcc = np.mean(lbr.feature.mfcc(y=wave, sr=sr, n_mfcc=n_mfcc), axis=1)
    music_features[i, : ] = mfcc
    
```

```python
# speech sound
path = "speech_data_final_750"
sph_files = os.listdir(path)
speech_features = np.zeros((len(sph_files), n_mfcc))

for i, fname in enumerate(sph_files):
    wave,sr = lbr.load(os.path.join(path, fname))
    mfcc = np.mean(lbr.feature.mfcc(y=wave, sr=sr, n_mfcc=n_mfcc), axis=1)
    speech_features[i, : ] = mfcc
    print(i)
```

```python
# nature sound
path = "nature_data_final_750"
nat_files = os.listdir(path)
nature_features = np.zeros((len(nat_files), n_mfcc))

for i, fname in enumerate(nat_files):
    wave,sr = lbr.load(os.path.join(path, fname))
    mfcc = np.mean(lbr.feature.mfcc(y=wave, sr=sr, n_mfcc=n_mfcc), axis=1)
    nature_features[i, : ] = mfcc
    print(i)
```

```python
all_files = np.vstack((machine_features, music_features, speech_features, nature_features))
lab_mach = np.zeros((len(mach_files),1), dtype = np.int8)
lab_mus = np.ones((len(mus_files),1), dtype=np.int8)
lab_sph = 2 * np.ones((len(sph_files),1), dtype=np.int8)
lab_nat = 3 * np.ones((len(nat_files),1), dtype=np.int8)

labels = np.vstack((lab_mach, lab_mus, lab_sph, lab_nat))
len_data = len(labels)
```

```python
print(labels.shape)
print(all_files.shape)
print(labels[4])
print(all_files[4,:])
```
