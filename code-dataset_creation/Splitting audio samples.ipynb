```python
# music sound
myaudio = AudioSegment.from_file("raw_data/finalfinalmusic.wav" , "wav") 
chunk_length_ms = 1000 # pydub calculates in millisec
chunks = make_chunks(myaudio, chunk_length_ms) #Make chunks of one sec
for i, chunk in enumerate(chunks):
    chunk_name = "music_data_final_1000/train_music{0}.wav".format(i)
    print ("exporting", chunk_name)
    chunk.export(chunk_name, format="wav")
```

```python
# speech sound
myaudio = AudioSegment.from_file("raw_data/speech_sound.wav" , "wav") 
chunk_length_ms = 1000 # pydub calculates in millisec
chunks = make_chunks(myaudio, chunk_length_ms) #Make chunks of one sec

for i, chunk in enumerate(chunks):
    chunk_name = "speech_data_final_1000/train_speech{0}.wav".format(i)
    print ("exporting", chunk_name)
    chunk.export(chunk_name, format="wav")
```

```python
# nature sound
myaudio = AudioSegment.from_file("raw_data/finalfinalnature.wav" , "wav") 
chunk_length_ms = 1000 # pydub calculates in millisec
chunks = make_chunks(myaudio, chunk_length_ms) #Make chunks of one sec

for i, chunk in enumerate(chunks):
    chunk_name = "nature_data_final_1000/train_nature{0}.wav".format(i)
    print ("exporting", chunk_name)
    chunk.export(chunk_name, format="wav")
```
